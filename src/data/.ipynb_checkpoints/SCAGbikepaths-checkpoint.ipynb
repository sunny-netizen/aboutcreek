# Ballona Creek Bike Path

```python
# subset ballona creek bike paths
creekpath = bikepaths[bikepaths.NAME=="BALLONA CREEK"].to_crs(3310)
creekpath.plot();
creekpath.head()
```

## Bike Path Connecting Roads 

```python
# Visually select IDs of intersecting bike path entrance roads; total of 13 
connections_ids = [
    36357, 36358, 36785, # National/Jefferson
    12216, 34613, # higuera
    12227, 19577, # duquesne
    19578, 
    33865, 27020, 27022, 33871, 34701, 28945, # sepulveda
    4567, 
    36542, 
    14540, 
    28671, 34677, 28735, 12635, #centinela
    26899, 5019, # McConnell
    13826, # Lincoln
    5130, # Fiji
    36676, # bridge
    36668 # Playa
]

# subset these ids from bikepaths
connections = bikepaths[bikepaths['OBJECTID'].isin(connections_ids)].to_crs(3310)

# check
ax = creekpath.plot();
connections.plot(ax=ax, color='red');
connections.head()
```

```python
groups = connections.dissolve().buffer(50)
groups = gpd.GeoDataFrame(geometry=groups).explode(index_parts=True).reset_index()
groups = groups[['geometry']]
groups
```

```python
connections = connections.sjoin(groups, how = 'left')
connections.head()
```

```python
connections = connections.dissolve(by='index_right')
connections.shape, connections.crs
```

## Bike Path Entrances

```python
from shapely.ops import nearest_points
entrances = [] 
for i, row in connections.iterrows():
    entrance = nearest_points(row['geometry'], creekpath.unary_union)[1]
    entrances.append(entrance)
entrances = gpd.GeoDataFrame(geometry=entrances, crs=3310)
entrances
```

```python
ax = creekpath.plot();
connections.plot(ax=ax, );
entrances.plot(ax=ax, color='red');
```

## Extent

```python
# we don't want to show all connection roads
ax = creekpath.plot();
connections.iloc[[10,12]].plot(ax=ax, color='red');
```

```python
# combine the crossroads we want to show with the creek bike path
creekpath2 = pd.concat([creekpath, connections.iloc[[10,12]]])
creekpath2.plot();
```

```python
# create a buffer based on the original bike path
creekbuffer = creekpath.buffer(1000)
creekbuffer.plot();
```

```python
# use the buffer to shorten the long segment in Playa del Rey
creekpath2 = creekpath2.clip(creekbuffer)
ax = creekpath2.plot();
```
